# Partie 1: Utiliser 4 variables

```python
clust_data_4v= data[['Population difference (%)', 'Animal protein (%)',\
       'Food supply (kcal/capita/day)','Protein supply quantity (g/capita/day)']].copy()
# data.drop(['Area Code', 'Year','Country Name','Country Code', 'GDP 2017', 'Region'], axis=1)
# préparation des données pour le clustering
X = clust_data_4v.values
names = data.index
features = clust_data_4v.columns

# Centrage et Réduction
std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)
```

## Dendrogram

```python
# Générer la matrice des liens
Z = linkage(X_scaled, 'ward')
```

```python
# # Dendrogramme tronqué 
plt.figure(figsize=(20,10))
plt.title('Hierarchical Clustering Dendrogram - Truncat')
plt.xlabel('distance')
plt.ylabel('Cluster size')
dendrogram(
    Z,
    truncate_mode='lastp',
    p=20,
    leaf_rotation=90.,
    leaf_font_size=12.,
    orientation='left',
#     show_contracted=True,
    color_threshold = 6.75,
)
plt.axvline(x=6.75, c='black')
plt.savefig('Graphes\\Dendogram_Truncat.jpg')
plt.show()
```

```python
# Dendrogram complet
plt.figure(figsize=(25,10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('distance')
dendrogram(
    Z,
    labels = names,
#     orientation = "left",
    color_threshold = 6.75,
    leaf_font_size = 8    
)
plt.axhline(y=6.75, c='black')
plt.savefig('Graphes\\Dendogram.jpg')
plt.show()
```
