### Mesure de forme pour chaque catégorie

```python
for cat in df_products["categ"].unique():
    subset = df_products.loc[df_products.categ == cat, :] # Création du sous-échantillon
    print("-"*20,"catégorie : ",cat,"-"*20)
    y1 = subset['price'].skew()
    print("Skewness empérique : ",y1)
    if (y1 == 0):
        print("--> Distribution symétrique")
    elif (y1 > 0):
        print("--> Distribution étalée à droite")
    else:
        print("Distribution étalée à gauche")
        
    y2 = subset['price'].kurtosis()
    print("kurtosis empérique : ",y2)
    if (y2 == 0):
        print("--> Même aplatissement que la distribution normale")
    elif (y2 > 0):
        print("--> moins aplatie que la distribution normale (observations plus concentrées)")
    else:
        print("--> plus aplatie que la distribution normale (observations moins concentrées)")
    
    print('\n*** Courbe de Lorenz ***')

    mf.courbe_lorenz(subset,'price',cat)
```

### Test Khi2

hypothèse nulle H0 : On suppose que le sexe des clients ne dépends pas des catégories du produit acheté <br/>

hypothèse à vérifier H1: Le sexe des clients dépend des catégories du produit acheté

```python
st_chi2, st_p, st_dof, st_exp = st.chi2_contingency(cont_table , correction = True)

# Statistique khi2, 
print('st_chi2 =', st_chi2)
 
# Nombe de degrés de liberté
print('Degré de liberté =', st_dof)
 
# Pvalue : l'unique valeur nécessaire pour prendre notre décision
print('p-value = .2f',round(st_p,20))
```

```python
# Test Khi2 selon l'axe 
st_chi2, st_p = st.chisquare(cont_table , ddof=0 , axis=0)

# Statistique khi2, 
print('st_chi2 =', st_chi2)
 
# Pvalue 
print(f'p-value = ',st_p)
```
