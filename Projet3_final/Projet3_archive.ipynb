###  Solution : changer les " éléments " en colonnes puis garder uniquement 645 et 664 
       

### A. Supprimer les Nan avant de faire mon opération

```python
df_rapport_calorique = pd.pivot_table (df_Alimentation, values='Valeur', index = ['Code zone','Code Produit'], 
                                columns = 'Code Élément')[[645,664]]
#df_rapport_calorique.dropna(inplace=True)
df_rapport_calorique = df_rapport_calorique[~df_rapport_calorique.isin([np.nan, np.inf, -np.inf, 0]).any(1)]
df_rapport_calorique['Ratio Energie/poids'] = df_rapport_calorique[664]*365 / df_rapport_calorique[645]
df_rapport_calorique['Unité du ratio'] ='kcal/kg'
df_rapport_calorique.head()
```

### B. Sans supprimer les Nan avant de faire mon opération

```python
df_rapport_calorique2 = pd.pivot_table (df_Alimentation, values='Valeur', index = ['Code zone','Code Produit'], 
                                columns = 'Code Élément')[[645,664]]
df_rapport_calorique2['Ratio Energie/poids'] = df_rapport_calorique2[664]*365 / df_rapport_calorique2[645]
df_rapport_calorique2['Unité du ratio'] ='kcal/kg'
df_rapport_calorique2.head(5)
```
